```python
from stockstats import stockstats
import numpy as np
```

```python
stock=stockstats(['TSLA','AAPL','FB','MSFT','AMC'],'2020-8-1','2020-9-1')
```

```python
hv=stock.HV()['AAPL'][0]
rshv=stock.RSHV()['AAPL'][0]
yzhv=stock.YZHV()['AAPL'][0]

print('\n     Volatility Estimator   ')
print('         Daily Results      ')
print('-----------------------------')
print('|Close-to-Close  :',round(hv,6)*100,'% |')
print('|Rogers-Satchell :',round(rshv,5)*100,'%  |')
print('|Yang-Zhang      :',round(yzhv,5)*100,'%  |')
print('-----------------------------\n')
```
